```python

apartmentlinks = []

for i in range(1,10):

    soup = BeautifulSoup(scraper.get(f'https://www.propertyguru.com.my/property-for-sale/{i}?freetext=Kuala+Lumpur&property_type=N&property_type_code%5B0%5D=APT&property_type_code%5B1%5D=CONDO&property_type_code%5B2%5D=FLAT&region_code=MY14').content,'html.parser')

    apartmentlist = soup.find_all('h3',class_='h4 ellipsis text-transform-none')

    for apt in apartmentlist:
        for link in apt.find_all('a',class_='nav-link'):
            link_href = link['href']
            apartmentlinks.append(link_href)



```

```python
# testlink =f'https://www.propertyguru.com.my/property-listing/allevia-mont-kiara-for-sale-by-larry-wong-34134264'

condominiumlist =[]

for link in apartmentlinks:

        soup2 = BeautifulSoup(scraper.get(link).text)
        results = soup2.find('div',class_='contents contents-listing-detail')


        # Find the element containing the text we want    

        #Proptertyname
        try:
            Propertyname = results.find('h1',class_='h2 text-transform-none').text.strip()
        except:
            Propertyname ='n/a'


        #Address
        try:
            address = results.find('span',itemprop='streetAddress').text.strip()
        except:
            address ='n/a'

        #Price
        try:
            Price = results.find('span',class_='element-label price').text.strip()
        except:
            Price ='n/a'

        #PSF
        try:
            PSF = results.find('tr',class_='property-attr price-per-area').text[5:].strip()
        except:
            PSF ='n/a'    

        #PropertyType
        try:
            PropertyType= results.find('td',class_='value-block').text.strip()
        except:
            PropertyType ='n/a'    


        #Bedrooms
        try:
            Bedrooms = results.find('span',itemprop='numberOfRooms').text.strip()
        except:
            Bedrooms ='n/a'    

        #Bathrooms
        try:
            Bathrooms = results.find('div',class_='property-info-element baths').text.strip()
        except:
            Bathrooms ='n/a'  

        #Floorsize
        try:
            Fr = results.find('tr',class_='floor-area')
        except:
            Fr = 'n/a'
        try:
            Floorsize = Fr.find('td',itemprop='value').text.strip()
        except:
            Floorsize ='n/a'  

        #Floorlevel
        try:
            Fl = results.find('tr',class_='floor-level')
        except:
            Fl = 'n/a'
        try:
            Floorlevel = Fl.find('td',class_='value-block').text.strip()
        except:
            Floorlevel ='n/a'

        #Furnishing
        try:
            Fur_nish = results.find('tr',class_='furnishing')
        except:
            Fur_nish ='n/a'
        try:
            Furnishing = Fur_nish.find('td',itemprop='value').text.strip()
        except:
            Furnishing ='n/a'


#             #Tenure 
#         try:
#             ten_ure = results.find('table',class_='row table-row').find_all('tbody')[6]
#         except:
#             ten_ure ='n/a'
#         try:
#             Tenure  = ten_ure.find('td',itemprop='value').text.strip()
#         except:
#             Tenure  ='n/a'

#         #Developer
#         try:
#             Dev = results.find('table',class_='row table-row').find_all('tbody')[2]
#         except:
#             Dev = 'n/a'
#         try:
#             Developer  = Dev.find('td',itemprop='value').text.strip()
#         except:
#             Developer  ='n/a'

#         #Completetion Year
#         try:
#             Year = results.find('table',class_='row table-row').find_all('tbody')[7]
#         except:
#             Year = 'n/a'
#         try:
#             Completetion_Year  = Year.find('td',itemprop='value').text.strip()
#         except:
#             Completetion_Year  ='n/a'

        unit_info ={
        'Propertyname' :Propertyname,
        'Address': address,
        'PropertyType' : PropertyType,
        'Price' : Price,    
        'PSF' : PSF,
        'Floorsize' : Floorsize,   
        'Floorlevel' : Floorlevel,
        'Furnishing' : Furnishing,
        'Bedrooms' : Bedrooms,
        'Bathrooms': Bathrooms       
        }    
            
        condominiumlist.append(unit_info)
    




            
```
